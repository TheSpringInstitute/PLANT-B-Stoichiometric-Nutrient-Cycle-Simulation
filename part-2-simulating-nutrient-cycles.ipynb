```python
# Create a directory to store the species stoichiometry data
species_stoichiometry_dir = os.path.join(stoichiometry_data_dir, 'species_stoichiometry_dir')
os.makedirs(species_stoichiometry_dir, exist_ok=True)
```

```python
# Create a directory to store the StoichLife data inside the species stoichiometry directory
StoichLife_dir = os.path.join(species_stoichiometry_dir, 'isopod_and_springtail_StoichLife_data')
os.makedirs(StoichLife_dir, exist_ok=True)

# Establish the URL to the XLSX file for the download
StoichLife_url = "https://datadryad.org/downloads/file_stream/3918707"

# Define the path to the XLSX file
StoichLife_path = os.path.join(StoichLife_dir, "datapaper_2025_02_11.xlsx")

"""Function code to successfully download the XLSX file is pending."""
```

```python
# Save the "data" sheet of the XLSX file as a dataframe
"""The XLSX file also contains a "metadata" and "references" sheet. The "metadata" sheet details what the variables (ex. habitat) are
describing."""
StoichLife_df = pd.read_excel(StoichLife_path, sheet_name="data")

# Display the data
StoichLife_df
```

```python
# Look at the columns
StoichLife_df.columns
```

```python
# Specify the columns to keep
"""Only columns related to the species's species name and family name and nutrient data will be kept."""
columns_to_keep = [
'sp.morphosp_init', 'sp.morphosp_revised', 'family_init', 'taxa.c', 'taxa.n', 'taxa.p', 'taxa.cn', 'taxa.cp', 'taxa.np', 'taxa.mass.cn', 'taxa.mass.cp', 'taxa.mass.np', 
'body.weight.gr', 'temperature', 'radiation', 'N', 'P_ter', 'P_mar'
]

# Filter the dataset to keep only the necessary columns
StoichLife_filtered_df = StoichLife_df[columns_to_keep]

# Display the data
StoichLife_filtered_df
```

```python
# Filter 'family_init' for a close relative of the dwarf white isopod
isopod_df = StoichLife_filtered_df[StoichLife_filtered_df['family_init'] == 'Platyarthridae']

# Display the result
isopod_df
```

```python
# Filter 'sp.morphosp_revised' for Folsomia fimetaria

springtail_df = StoichLife_filtered_df[StoichLife_filtered_df['sp.morphosp_revised'] == 'Folsomia fimetaria']
                                       
# Display the results
springtail_df
```
